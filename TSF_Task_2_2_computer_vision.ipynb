###Implement an image color detector which identifies all the colors in an image.

###Colours are extracted from the image using KMeans Algorithm, OpenCv is used for image manipulation.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import cv2
from collections import Counter
from skimage.color import rgb2lab, deltaE_cie76
```

##Convert the RBG values in hexadecimal values

```python
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

##Kmeans Algorithm is used to Extract colors.KMeans algorithm creates clusters based on the supplied number_of_colors.

Then, the cluster colors are identified an arranged in the correct order. We plot the colors as a pie chart.

```python
def get_colors(image_path, number_of_colors, show_chart):
    
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.title("Image")
    plt.imshow(image)
    
    modified_image = image.reshape(image.shape[0]*image.shape[1], 3)
    
    clf = KMeans(n_clusters = number_of_colors)
    labels = clf.fit_predict(modified_image)
    
    counts = Counter(labels)
    # sort to ensure correct color percentage
    counts = dict(sorted(counts.items()))
    
    center_colors = clf.cluster_centers_
    # We get ordered colors by iterating through the keys
    ordered_colors = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered_colors[i]) for i in counts.keys()]
    rgb_colors = [ordered_colors[i] for i in counts.keys()]


    if (show_chart):
        plt.figure(figsize = (8, 6))
        plt.title("Detected colours")
        plt.pie(counts.values(), labels = hex_colors, colors = hex_colors)
```

```python
get_colors('sample_2.jfif', 6, True)
```
